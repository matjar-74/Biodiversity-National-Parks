# II. Data analysis

## 1. Importing and inspecting the data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
observations = pd.read_csv('observations.csv')
speciesInfo = pd.read_csv('species_info.csv')
```

```python
print('The \'observations\' database consists of ', observations.shape[1], ' columns and ', observations.shape[0], ' records.')
observations.head()
```

```python
print('National Parks included in the database are: ', observations.park_name.unique())
print('Total number of specimens observed: ', observations.observations.sum())
```

```python
print('The \'speciesInfo\' database consists of ', speciesInfo.shape[1], ' columns and ', speciesInfo.shape[0], ' records.')
print('Number of unique species: ', speciesInfo.scientific_name.nunique())
speciesInfo.head()
```

```python
print('Categories of taxonomy included in the database are: ', speciesInfo.category.unique())
print('Conservation statuses included in the database are: ', speciesInfo.conservation_status.unique())
```

Having 'NaN' values in a database always require special attention.
In this database the occurance of such value in a 'conservation_status' column makes perfect sense. The species without a value provided are simply not protected species. The missing data is **data structurally missing**.

After checking the distribution of statuses of conservation, we can see that there are a lot of species (161) which situation is starting to concern biologists. The lower numbers of 16 and 10 are endangered and threatened species respectively. The lowest number of 4 are the recovering species.

```python
speciesInfo.groupby('conservation_status').common_names.count().reset_index().rename(columns={'conservation_status': 'Status of Conservation', 'common_names':'Number'})
```

## 2. Analysis of the categories of taxonomy

Checking the categories of species in the database shows that an overwhelming majority of the recorded species (about 77%) are of the Vascular Plant category. The rest of the categories take up more compareable proportions of the population ranging from Birds (9%) to Reptiles (1.4%).

The distribution across the categories changes significantly when we take into consideration only the species OSLOC. Here we find out that (while being the wast majority of the species in the database) the Vascular Plants take the second place behind Birds. The third most numerous group of Nonvascular Plants end up at the last place. Apart from those two outliers all the other categories stay in the same order with fairly logical proportions.

Such characteristics visible on the charts below suggest that **plants are either very resiliant to becoming an endangered species or their situation is less precisely explored** by the experts.

```python
# Creating a color map to have consistent colors for each category.
categoriesColorMap = {'Mammal': '#fcad03',
                      'Bird': '#6ca3e6',
                      'Reptile': '#663f1e',
                      'Amphibian': '#48ab7d',
                      'Fish': '#07338c',
                      'Vascular Plant': '#26bd26',
                      'Nonvascular Plant': '#4d6316'}
```

```python
speciesInfo['concerning'] = speciesInfo.conservation_status.notna()
```

```python
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.barplot(x = 'category', y = 'count', data=speciesInfo.category.value_counts(ascending=False).reset_index(),
            hue='category', palette=categoriesColorMap)
plt.xlabel('Category of the species')
plt.ylabel('Number of species')
plt.title('Number of recorded species \n(per category)')
plt.xticks(rotation=30)

plt.subplot(1,2,2)
sns.barplot(x = 'category', y = 'count', data=speciesInfo[speciesInfo['concerning']==True].category.value_counts(ascending=False).reset_index(),
            hue='category', palette=categoriesColorMap)
plt.xlabel('Category of the species')
plt.ylabel('Number of species')
plt.title('Number of recorded species OSLOC \n(per category)')
plt.xticks(rotation=30)

plt.show()
plt.close()
```

Checking the **percentage of species being OSLOC** for each category will also give us some more insight into given database.

As expected the table and the chart below show that both categories of Plants have a very small percentage score of 1-1.5% of OSLOC species. The Mammals and Birds have the highest values of almost 18% and 15% respectively, while Amphibians, Fishes and Reptiles - between 6% and 9%

```python
total_counts = speciesInfo.groupby('category')['scientific_name'].count()
endangered_counts = speciesInfo[speciesInfo['concerning']==True].groupby('category')['scientific_name'].count()
endangeredPercent = (endangered_counts / total_counts * 100).sort_values(ascending=False).reset_index()
endangeredPercent.rename({'category':'Category of Taxonomy', 'scientific_name':'Percentage of species OSLOC'})
print(endangeredPercent)

plt.figure(figsize=(6,4))
sns.barplot(x=endangeredPercent.category, y=endangeredPercent.scientific_name, hue=endangeredPercent.category, palette=categoriesColorMap)
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlabel('Category of species')
plt.ylabel('Percentage of endangered species')
plt.title('Percentage of species endangered to some level (per category)')
plt.show()
plt.close()
```

## 3. Analysis of the statuses of concern

As already mentioned, there are 5 levels of concern considered in the database:
- Not Concerning,
- Species of Concern,
- Threatened,
- Endangered,
- In Recovery.

As quantity of the species sum up to over 5.5 thousands, number of species OSLOC have only 190 positions on the list, which is less than 3.5%. Luckily the situation of huge majority of the recorded species is not alarming. If we narrow down the research to only Endangered species, we are left with 16 species, that need people's support.

By looking at the distribution of the species' categories across statuses we can find insightfull trends.

Firstly, while Birds are the most represented category among the species OSLOC, a huge majority of the records belong to the Species of Concern status. Their numbers in the Threatened or Endangered are fairly low. This means that while there are a lot of them that deserve our attention, **the Birds' situation is not as bad as it would seem in the beginning**.

Secondly, the **In Recovery status consists of Mammals and Birds only**. It's logical that they should be the majority of records here, as they're the most endangered ones, but the fact that there are no other categories represented is quite alarming. It suggests that our efforts to protect the endangered species are effective exclusively towards Birds and Mammals. We have not yet found a successful way of protecting other categories of plants and animals.

Lastly, **the most resilient Categories of species seem to be Reptiles and Nonvascular Plants**. Assuming that the data is accurate, these show up only in the Species of Concern status and in low quantities.

```python
print('Number of all the species: ', speciesInfo.common_names.nunique())
print('Number of species OSLOC: ', speciesInfo[speciesInfo.concerning==True].common_names.nunique())
print('Number of Endangered species: ', speciesInfo[speciesInfo.conservation_status=='Endangered'].common_names.nunique())
```

## 4. Analysis of the observations in National Parks

Thanks to the `observations` database we can inspect observed number of specimens of each species in each of the four National Parks.
It is necessary to combine this data with data from `speciesInfo` database to get the information about Status of Concern of each species.

We've already found out that there are over 3.3 mil specimens observed over the last week. Checking their distribution between parks gives us some idea of the size of each park. Comparing these numbers to the counts of specimens belonging to the species OSLOC show that their proportions are kept almost identically.
The distribution of specimens across categories of taxonomy does also look practically the same for each park, only the scale of the numbers shows the difference between them. It shows that these parks have very similar ecosystems.

```python
parkAnalysis = pd.merge(observations, speciesInfo)
```

```python
plt.figure(figsize=(12,4))

ax = plt.subplot(1,2,1)
sns.barplot(x='park_name', y='observations', data=parkAnalysis.groupby('park_name').observations.sum().reset_index(), color = 'green')
plt.xticks(rotation=30)
ax.set_xticks(range(4))
ax.set_xticklabels(['Bryce National Park','Great Smoky Mountains\nNational Park','Yellowstone National Park','Yosemite National Park'])
plt.xlabel('Park Name')
plt.ylabel('Number of observed specimens')
plt.title('Number of all the specimens\nobserved across National Parks')

ax = plt.subplot(1,2,2)
sns.barplot(x='park_name', y='observations', data=parkAnalysis.dropna().groupby('park_name').observations.sum().reset_index(), color = 'orange')
plt.xticks(rotation=30)
ax.set_xticks(range(4))
ax.set_xticklabels(['Bryce National Park','Great Smoky Mountains\nNational Park','Yellowstone National Park','Yosemite National Park'])
plt.xlabel('Park Name')
plt.ylabel('Number of observed specimens')
plt.title('Number of recorded specimens\nthat are endangered to some level')

plt.show()
plt.close()

plt.figure(figsize=(12,4))
sns.barplot(x='park_name', y='observations', data=parkAnalysis.groupby(['park_name','category']).observations.sum().reset_index(), hue='category', palette=categoriesColorMap)
plt.xticks(rotation=30)
plt.title('Distribution of specimens across categories')
plt.legend(title='Categories')
plt.xlabel('Park Name')
plt.ylabel('Number of observed specimens')
plt.grid(axis='y', linestyle='--')

plt.show()
plt.close()
```

It's time to check on the numbers of the species in the worst situation - the Endangered species.

The table below shows the **species with the lowest amount of specimens** for each category in every park. It shows which species require attention most urgently. Remembering that the Amphibian category has only one Endangered species, we are able not to overreact seeing Sierra Nevada Yellow-Legged Frog showing up for every single park.

Majority of the numbers are proportional to the overall sizes of the individual parks population wise. The highlights are the quantities in Fish and Vascular Plants categories in the Great Smoky Mountains National Parks, which even surpass the numbers in Bryce National Park. It would be a good idea to inspect the practices implemented in this park and look for possible causes for this success.

```python
maxEndangered = parkAnalysis.loc[parkAnalysis[parkAnalysis.conservation_status == 'Endangered']\
.groupby(['park_name','category']).observations.idxmin(), ['park_name','category','observations', 'common_names']]
maxEndangered['pivotValue']= maxEndangered.common_names.str.cat(maxEndangered.observations.astype(str), sep=": ")
maxEndangered.pivot(index='park_name', columns='category', values='pivotValue')
```

Lastly, analysing the data for a **species with lowest number of observations** including all the parks is Whooping Crane. This species has only 131 animals in the parks recorded during the week of collecting data.

```python
minObservations = parkAnalysis\
.groupby(['common_names']).observations.sum().reset_index().sort_values(by='observations')
print(minObservations.iloc[0])
```

# III. Conclusion

This project was able to analyse provided data, draw conclusions from the analysis and visualize some of the most important subjects. Some trends were found associated with the categories, statuses of concern and National Parks included in the database.

These are the questions asked in the beginning of this document with brief answers to them:
- *How many species are currently protected?*  
        Out of the 5504 species in the database there are 190 species that are of some level of concern.
- *What is the distribution of all observed species across categories? How is it different for the species OSLOC?* 
        The most numerous category is the Vascular Plants category, which makes up about 75% of all species. After that there are Birds, Nonvascular Plants, Mammals, Fishes, Amphibians, Reptiles ranging from 9% to 1.4%. The distribution of species OSLOC show that plants are less likely to be endangered than animals.
- *Which category of species shows the greatest likelihood of being species OSLOC?*  
        Species from Mammals category are most likely to be OSLOC, as 17.8% of all Mammal species already belong to that group.
- *What is the distribution of categories for each status of conservation?*  
        The most prevelant status of Species of Concern consists mostly of Birds, Vascular Plants and Mammals. It is also the only status that has species from every single category. In Recovery status has only a couple of species of Birds and Mammals. The Threatened and Endangered statuses have diverse compositions.
- *How successful is protecting the species? What species are recovering?*  
        Only the species from Birds and Mammals categories are currently protected effectively, as they are considered to be recovering.
- *What are the species that require our attention most urgently?*  
        The table included in the document shows most endangered species for each category and each park, but overall the most worrying is the situation of Whooping Crane, which is represented by only 131 specimens across these four National Parks.
